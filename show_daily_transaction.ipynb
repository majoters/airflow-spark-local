```python
from pyspark.sql import SparkSession

spark = SparkSession \
    .builder \
    .appName("spark-nb") \
    .master("spark://spark-master:7077") \
    .enableHiveSupport() \
    .getOrCreate()
```

```python
daily_transaction_df = spark.sql("SELECT * FROM local_db.daily_transaction")
```

```python
daily_transaction_df.toPandas()
```

```python
sorted_transaction_df = daily_transaction_df.orderBy("transaction_date")
```

```python
sorted_transaction_df.toPandas()
```

```python
result_path = "s3a://spark-warehouse/raw_data/save_from_notebook"
sorted_transaction_df.write \
    .format("csv") \
    .option("header", "true") \
    .mode("overwrite") \
    .save(result_path)
```

```python
spark.stop()
```
